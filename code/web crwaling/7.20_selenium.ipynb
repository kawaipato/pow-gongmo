```python
import selenium #셀레니움
import time
import pandas as pd #csv를 읽고 dataframe을 사용하기 위한 pandas
from selenium import webdriver #브라우저를 띄우고 컨트롤하기 위한 webdriver
from selenium.webdriver.common.keys import Keys #브라우저에 키입력 용
from selenium.webdriver.common.by import By #webdriver를 이용해 태그를 찾기 위함
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait #Explicitly wait을 위함
from webdriver_manager.chrome import ChromeDriverManager #크롬에서 크롤링 진행 크롬 웹 드라이버 설치시 필요
from selenium.webdriver.support import expected_conditions as EC #브라우저에 특정 요소 상태 확인을 위해
from bs4 import BeautifulSoup #브라우저 태그를 가져오고 파싱하기 위함
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,TimeoutException #예외처리를 위한 예외들 
```

```python
# driver 실행
driver = webdriver.Chrome(service= Service(ChromeDriverManager().install())) 
# kakao map 접속
driver.get("https://map.kakao.com/") 
keyword='반려동물동반'
# 반려동물동반 키워드 검색
kakao_map_search_url = f"https://map.kakao.com/?q={keyword}" 
driver.get(kakao_map_search_url)
# 클릭 시간 여유 보장!
driver.implicitly_wait(time_to_wait=3)

# 1번째 리스트 만들기
jj = []
for i in range(1,16):
    nam = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.head_item.clickArea > strong > a.link_name'.format(i)).text
    typ = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.head_item.clickArea > span'.format(i)).text
    add = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.info_item > div.addr > p:nth-child(1)'.format(i)).text
    tim = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.info_item > div.openhour > p > a'.format(i)).text
    jj.append([nam,typ,add,tim])
more_btn = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.more')
more_btn.click()
time.sleep(.3)
```

```python
# 2~5번째 리스트 생성
for j in range(2,6):
    yoho = driver.find_element(By.CSS_SELECTOR,'#info\.search\.page\.no{}'.format(j)) # 다음장 클릭
    yoho.click()
    time.sleep(.3)
    for i in range(1,16):
        nam = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.head_item.clickArea > strong > a.link_name'.format(i)).text
        typ = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.head_item.clickArea > span'.format(i)).text
        add = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.info_item > div.addr > p:nth-child(1)'.format(i)).text
        tim = driver.find_element(By.CSS_SELECTOR,'#info\.search\.place\.list > li:nth-child({}) > div.info_item > div.openhour > p > a'.format(i)).text
        jj.append([nam,typ,add,tim])
next_btn = driver.find_element(By.CSS_SELECTOR,'#info\.search\.page\.next')
next_btn.click()
time.sleep(.3)
```
