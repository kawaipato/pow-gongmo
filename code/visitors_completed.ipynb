```python
import pandas as pd
```

```python
df1 = pd.read_csv('20230726212653_광역별 방문자 수.csv',encoding='CP949')
df2 = pd.read_csv('final_animal_visit_travel2.csv',encoding='cp949')
```

```python
display(df1.head())
display(df2.head())
```

```python
len(df2), len(df1)
```

```python
df1['시군구'] = df1['기초지자체명'].str[:4].str.strip()
```

```python
df1.loc[df1['광역지자체명'].str.contains('광역시'),'시군구'] = df1[df1['광역지자체명'].str.contains('광역시')]['광역지자체명'] + ' ' + df1[df1['광역지자체명'].str.contains('광역시')]['기초지자체명']
```

```python
df1[df1['광역지자체명'].str.contains('광역시')]
```

```python
df1.loc[df1.기초지자체명 == '세종특별자치시', '시군구'] = '세종시'
```

```python
df1['시군구'].unique()
```

```python
df3 = df1.groupby('시군구',as_index=False)['기초지자체 방문자 수'].sum()
```

```python
df = pd.merge(df2,df3,on='시군구',how='left')
```

```python
df.to_csv('final_animal_visit_travel3.csv',encoding='cp949',index=False)
```
