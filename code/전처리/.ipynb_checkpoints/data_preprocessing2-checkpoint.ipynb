```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 데이터 불러오기
df = pd.read_csv('all_data_plus.csv',encoding='cp949')
df
```

```python
# 고성군 중복값 처리
df.loc[(df.si == '고성군') & (df.do == '경남'),'시군구(특별)'] = '경남 고성군'
df.loc[(df.si == '고성군') & (df.do == '강원특별자치도'),'시군구(특별)'] = '강원도 고성군'
```

```python
# 데이터 저장
df.to_csv('all_data_plus.csv',encoding='cp949',index=False)
```

```python
# 카테고리 중, 개수가 적은 카테고리 묶기
dd = df.copy()
dd.loc[df['category'].isin(['문화시설','병원','약국','주차장']),'category'] = '기타'
```

```python
# 시군구에 따른 개수 빈도표
df1 = pd.crosstab(index=dd['시군구(특별)'],columns=dd['category'])
```
