```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
os.getcwd()
```

```python
# 해당 raw 파일들 불러오기
file_list = os.listdir(os.getcwd())
file_list_py = [file for file in file_list if file.endswith('.csv')]
file_list_py
```

```python
# 파일 경로와 변수 이름을 매핑한 딕셔너리
file_variable_map = {
    'kakao_map_animal_애견동반.csv': 'animal_dongban_data',
    'kakao_map_animal_반려동물공원.csv': 'pet_park_data',
    'kakao_map_animal_애견동반식당.csv': 'animal_restaurant_data',
    'kakao_map_animal_raw2.csv': 'raw_data2',
    'kakao_map_animal_raw3.csv': 'raw_data3',
    'kakao_map_animal_애견동반숙소.csv': 'animal_accommodation_data',
    'kakao_map_animal_애견동반카페.csv': 'animal_cafe_data',
    'kakao_map_animal_산책로.csv': 'walking_trail_data',
    'kakao_map_animal_raw.csv': 'raw_data',
    'kakao_map_animal_반려견동반.csv': 'pet_dongban_data'
}

# 각 파일을 불러와서 변수에 저장
for file_path, variable_name in file_variable_map.items():
    df = pd.read_csv(file_path,encoding='cp949')
    globals()[variable_name] = df
```

```python
# 파일들을 담을 리스트
data_frames = []

# 각 파일을 DataFrame으로 불러와서 리스트에 추가
for file in file_list_py:
    df = pd.read_csv(file,encoding='cp949')
    data_frames.append(df)

# 모든 데이터를 행 방향으로 합치기
combined_data = pd.concat(data_frames, ignore_index=True)

# 합쳐진 데이터 확인
all_data = pd.DataFrame(combined_data)
all_data
```

```python
# 데이터 na값 파악
all_data.isnull().sum()
```

```python
# 주소, 이름 중복값 제거
all_data_drop = all_data.drop_duplicates(['addr','name'],keep='first')
all_data_drop
```

```python
# drop한 데이터프레임 확인
all_data_drop
```

```python
# 병합한 데이터 저장
all_data_drop.to_csv('animal_data.csv',encoding='cp949',index=False)
```

```python
# drop한 데이터프레임 na값 확인
all_data_drop.isnull().sum()
```

```python
# 카테고리에 포함되어있지 않은 경우 1638개 존재하여 카테고리 기타로 추가함
all_data_drop[all_data_drop.cat_code.isnull()]
```

```python
# null값이 포함된 카테고리 확인
all_data_drop[all_data_drop.cat_code.isnull()].category.unique()
```

```python
# 범주형 코드 리스트 unique 값 저장
cat_cod_lst = all_data_drop.cat_code.unique()
cat_cod_lst
```

```python
# cateogory code, category 변수의 매치에 문제가 없는 걸로 보임!
cat_lst = []
for i in cat_cod_lst:
    cat_lst.append(all_data_drop[all_data_drop.cat_code == i].category.unique())
cat_lst
```

```python
# 카테고리별 빈도수 확인
category_lst = all_data_drop.groupby('category').size().sort_values(ascending=False)
category_lst
```

### Ex) 충남 -> 충청남도 로 바꾸는 작업

```python
state_lst = {'경남':'경상남도',
            '경북' : '경상북도',
            '전남':'전라남도',
            '전북':'전라북도',
            '충남':'충청남도',
            '충북':'충청북도',
            '경기':'경기도',
            '서울':'서울특별시',
            '광주':'광주광역시',
            '울산':'울산광역시',
            '대구':'대구광역시',
            '대전':'대전광역시',
            '부산':'부산광역시',
             '인천':'인천광역시'
            }
all_data_drop['시도']=all_data_drop['do'].replace(state_lst)
```

```python
all_data_drop['시군구'] = all_data_drop['si']
# all_data_drop['시도'] = all_data_drop['dosi']
# all_data_drop.drop(columns=['do','si','gu','dosi'],axis=1,inplace=Tru
```

```python
# 반려견 동반 출입가능 장소 카테고리별 그래프 작성
palette = sns.color_palette('Set2')
sns.barplot(y=category_lst.index,x=category_lst,palette=palette)
plt.title('2023\' 반려견 동반 출입가능장소 종류')
plt.ylabel('')
plt.show()
```

```python
# 전국 시군구별 반려동물 동반 가능 장소 barplot 작성
dosi = all_data_drop.groupby('시도').size().sort_values(ascending=False)
sns.barplot(y=dosi.index,x=dosi,palette=palette)
plt.title('2023\' 전국 반려동물 동반 가능 장소')
plt.ylabel('')
plt.show()
```

```python
# 시군구 중복 열 처리
all_data_drop['시군구(특별)'] = all_data_drop['시군구']
all_data_drop.reset_index(inplace=True,drop=True)
```

```python
# 최종 데이터프레임 확인
all_data_drop
```

```python
# 세종시 및 광역시 시군구 열 처리
df = all_data_drop.copy()
df.loc[df['시도'] == '세종특별자치시', '시군구(특별)'] = '세종시'
df.loc[df['시도'].isin(['부산광역시','인천광역시','광주광역시','대구광역시','대전광역시','울산광역시']),'시군구(특별)'] = df.loc[df['시도'].isin(['부산광역시','인천광역시','광주광역시','대구광역시','대전광역시','울산광역시'])]['시도'] + ' ' + df.loc[df['시도'].isin(['부산광역시','인천광역시','광주광역시','대구광역시','대전광역시','울산광역시'])]['시군구']
```

```python
# 전국 시군구 반려동물 동반 가능 장소 가장 많은 도시 선정[시/군구/ 중 (시)로 구분]
sigun = df.groupby('시군구(특별)').size().sort_values(ascending=False).head(10)
sns.barplot(y=sigun.index,x=sigun,palette=palette)
plt.title('2023\' 전국 시군구 반려동물 동반 가능 장소')
plt.ylabel('')
plt.show()
```

```python
# 반려동물 동반지역 많은 도시 중 카테고리별 정보 확인
df10_idx = df.groupby('시군구(특별)').size().sort_values(ascending=False).head(10).index
df10 = df[df['시군구(특별)'].isin(df10_idx[:5])]
sns.catplot(x='시군구(특별)', kind='count', hue='category', data=df10, height=7, aspect=2)
plt.xlabel('시군구')
plt.ylabel('')
plt.title('반려동물 동반지역 TOP 5',size=)
plt.show()
```
