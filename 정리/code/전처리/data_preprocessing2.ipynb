```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('all_data_plus.csv',encoding='cp949')
df
```

```python
df.loc[(df.si == '고성군') & (df.do == '경남'),'시군구(특별)'] = '경남 고성군'
df.loc[(df.si == '고성군') & (df.do == '강원특별자치도'),'시군구(특별)'] = '강원도 고성군'
```

```python
df1 = pd.crosstab(index=dd['시군구(특별)'],columns=dd['category'])
```

## 방문자수 2022 데이터전처리

```python
df = pd.read_csv('방문자수_2022.csv',encoding='cp949')
df.head()
```

```python
df['시군구'] = df['기초지자체명']
```

```python
df.loc[df['광역지자체명'].str.contains('광역시'),'시군구'] = df[df['광역지자체명'].str.contains('광역시')]['광역지자체명'] + ' ' + df[df['광역지자체명'].str.contains('광역시')]['기초지자체명']
```

```python
df.loc[-df['광역지자체명'].str.contains('광역시'),'시군구'] = df.loc[-df['광역지자체명'].str.contains('광역시'),'시군구'].str[:3]
```

```python
df.to_csv('visitors.csv',encoding='cp949')
```
